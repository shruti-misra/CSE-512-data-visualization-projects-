```python
#Import relevant libraries 
import pandas as pd
import altair as alt
```

```python
#Import sunshine data from the Data folder
data = pd.read_csv('gdrive/My Drive/Grad School-Files/CSE 512/Data/sunshine.csv')
data
```

```python
#Look at the longitudinal information, which will come in handy later to create the visualization

longitude = [data[data['city'] == 'New York']['lon'].unique()[0], data[data['city'] == 'Miami']['lon'].unique()[0], 
             data[data['city'] == 'Chicago']['lon'].unique()[0], data[data['city'] == 'Houston']['lon'].unique()[0],
             data[data['city'] == 'Seattle']['lon'].unique()[0], data[data['city'] == 'San Francisco']['lon'].unique()[0]]
longitude
```

```python
#Categorize months into seasons

winter = ['Dec', 'Jan', 'Feb']
spring = ['Mar', 'Apr', 'May']
summer = ['Jun', 'Jul', 'Aug']
fall =   ['Sep', 'Oct', 'Nov']
```

```python
#Function to compute the average sunshine in a season 

def get_avg_sunshine(data, season):
  data_season = data[data["month"].isin(season)]
  return data_season['sunshine'].mean()
```

```python
#Compute average seasonal hours of sunshine for each of the 6 major cities

Seattle = data[data['city'] == 'Seattle']
Seattle_fall = get_avg_sunshine(Seattle, fall)
Seattle_win = get_avg_sunshine(Seattle, winter)
Seattle_spr = get_avg_sunshine(Seattle, spring)
Seattle_sum = get_avg_sunshine(Seattle, summer)

Chicago = data[data['city'] == 'Chicago']
Chicago_fall = get_avg_sunshine(Chicago, fall)
Chicago_win = get_avg_sunshine(Chicago, winter)
Chicago_spr = get_avg_sunshine(Chicago, spring)
Chicago_sum = get_avg_sunshine(Chicago, summer)

Houston = data[data['city'] == 'Houston']
Houston_fall = get_avg_sunshine(Houston, fall)
Houston_win = get_avg_sunshine(Houston, winter)
Houston_spr = get_avg_sunshine(Houston, spring)
Houston_sum = get_avg_sunshine(Houston, summer)

Miami = data[data['city'] == 'Miami']
Miami_fall = get_avg_sunshine(Miami, fall)
Miami_win = get_avg_sunshine(Miami, winter)
Miami_spr = get_avg_sunshine(Miami, spring)
Miami_sum = get_avg_sunshine(Miami, summer)

NY = data[data['city'] == 'New York']
NY_fall = get_avg_sunshine(NY, fall)
NY_win = get_avg_sunshine(NY, winter)
NY_spr = get_avg_sunshine(NY, spring)
NY_sum = get_avg_sunshine(NY, summer)

SF = data[data['city'] == 'San Francisco']
SF_fall = get_avg_sunshine(SF, fall)
SF_win = get_avg_sunshine(SF, winter)
SF_spr = get_avg_sunshine(SF, spring)
SF_sum = get_avg_sunshine(SF, summer)
```

```python
#Create a new dataframe with cities and seasons

seasons = pd.DataFrame({
    'city': ['Seattle', 'Seattle', 'Seattle', 'Seattle','Chicago', 'Chicago', 'Chicago', 'Chicago',
             'Miami', 'Miami', 'Miami', 'Miami','Houston', 'Houston', 'Houston', 'Houston',
             'New York', 'New York', 'New York', 'New York','San Francisco', 'San Francisco', 'San Francisco', 'San Francisco'],

    'season': ['Winter', 'Spring', 'Summer', 'Fall','Winter', 'Spring', 'Summer', 'Fall',
               'Winter', 'Spring', 'Summer', 'Fall', 'Winter', 'Spring', 'Summer', 'Fall', 
               'Winter', 'Spring', 'Summer', 'Fall', 'Winter', 'Spring', 'Summer', 'Fall'],

    'sunshine': [Seattle_win, Seattle_spr, Seattle_sum, Seattle_fall,
                 Chicago_win, Chicago_spr, Chicago_sum, Chicago_fall,
                 Miami_win, Miami_spr, Miami_sum, Miami_fall,
                 Houston_win, Houston_spr, Houston_sum, Houston_fall,
                 NY_win, NY_spr, NY_sum, NY_fall,
                 SF_win, SF_spr, SF_sum, SF_fall]
}) 
```

```python
#Round the average hours of sunshine to two decimal places for ease of visualization

seasons['sunshine'] = seasons['sunshine'].round(decimals = 1)
seasons
```

```python
# Visualization without text

heatmap = alt.Chart(seasons).mark_rect().encode(
    alt.X('season:N', sort=["winter", "spring", "summer", "fall"]),
    alt.Y('city:N', sort=["New York", "Miami", "Chicago", "Houston", "Seattle", "San Francisco"]),
    alt.Color('sunshine:Q', scale = alt.Scale(scheme = 'Inferno'), legend=alt.Legend(title="Average hours of sunshine"))
).properties(
    title = 'Average Seasonal Sunshine Patterns in 6 Major U.S Cities',
    width = 500,
    height = 500
).configure_axis(
    labelFontSize=15,
    titleFontSize=15
).configure_title(fontSize=15, anchor="middle").configure_legend(
    gradientLength=200,
    gradientThickness=30,
titleFontSize=10,
labelFontSize=10
)

heatmap
```

```python
# Visualization with text

heatmap = alt.Chart(seasons).mark_rect().encode(
    alt.X('season:N', sort=["winter", "spring", "summer", "fall"]),
    alt.Y('city:N'),
    alt.Color('sunshine:Q', scale = alt.Scale(scheme = 'Inferno'), legend=alt.Legend(title="Average hours of sunshine"))
)

text = heatmap.mark_text().encode(
    text = 'sunshine:Q',
    color=alt.condition(
        alt.datum.sunshine > 250,
        alt.value('black'),
        alt.value('white')
    )
) 

(heatmap + text).properties(
    title = 'Average Seasonal Sunshine Patterns in 6 Major U.S Cities',
    width = 500,
    height = 500
).configure_axis(
    labelFontSize=15,
    titleFontSize=15
).configure_title(fontSize=20, anchor="middle").configure_legend(
    gradientLength=200,
    gradientThickness=30,
titleFontSize=10,
labelFontSize=10
)
```
